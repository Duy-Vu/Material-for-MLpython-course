```python
import numpy as np
from sklearn import linear_model 
from sklearn.datasets import make_regression 

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import animation
```

More complicated landscape

 six-hump Camel function 
 The function is usually evaluated on the rectangle x1 ∈ [-3, 3], x2 ∈ [-2, 2]
 The function has six local minima, two of which are global. 
 
 https://www.sfu.ca/~ssurjano/camel6.html
 
 https://en.wikipedia.org/wiki/Test_functions_for_optimization

# Functions

```python
#--------------------------------#
# FUNCTIONS AND THEIR DERIVATIVES

# BOWL

def bowl(x,y):
    
    z = x**2 + y**2
    
    return z

def bowl_der(x,y):
        
    # Partial derivatives
    dx  = 2*x
    dy  = 2*y
    
    return dx, dy

# SADDLE

def saddle(x,y):
    
    z = x**2 - y**2
    
    return z

def saddle_der(x,y):
        
    # Partial derivatives
    dx  = 2*x
    dy  = -2*y
    
    return dx, dy

# CAMEL

def camel(x,y):
    
    term1 = (4-2.1*x**2+(x**4)/3)*x**2
    term2 = x*y
    term3 = (-4+4*y**2)*y**2
    z = term1 + term2 + term3
    
    return z

def camel_der(x,y):

    # Partial derivatives

    dt1_x = 8*x - 8.4*x**3 + 2*x**5
    dt2_x = y
    dt3_x = 0

    dx = dt1_x + dt2_x + dt3_x

    dt1_y = 0
    dt2_y = x
    dt3_y = -8*y + 16*y**3

    dy = dt1_y + dt2_y + dt3_y
    
    return dx, dy
```

# Optimizers

```python
#--------------------------------#
# OPTIMIZERS

def plain_gd(x,y,func,lrate,epoches):
    
    log = []
    for epoch in range(epoches):
        
        log.append([x,y])
        dx, dy = func(x,y)

        x   = x - lrate* dx        
        y   = y - lrate* dy         
        
    log = np.asarray(log)       
    return log  

def sgmoment(x,y,func,lrate,epoches,params):
    
    beta = params
    vx, vy = 0, 0
    
    log = []
    for epoch in range(epoches):
        
        log.append([x,y])
        dx, dy = func(x,y)  

        vx = beta*vx + (1-beta)*dx  
        vy = beta*vx + (1-beta)*dy

        x   = x - lrate*vx   
        y   = y - lrate*vy  

    log = np.asarray(log)      
    return log

def rmsprop(x,y,func,lrate,epoches,params):

    beta, eps = params
    vx, vy = 0, 0
    
    log = []
    for epoch in range(epoches):
        
        log.append([x,y])   
        dx, dy = func(x,y)    

        vx = beta*vx + (1-beta)*dx**2   
        vy = beta*vy + (1-beta)*dy**2 

        x   = x - (lrate/(np.sqrt(vx)+eps))* dx         
        y   = y - (lrate/(np.sqrt(vy)+eps))* dy          
    
    log = np.asarray(log)
    return log

def adam(x,y,func,lrate,epoches,params):
    
    beta1, beta2, eps = params
    vx, vy, sx, sy = 0, 0, 0, 0
    
    log = []
    for epoch in range(epoches):
        
        log.append([x,y])           
        dx, dy = func(x,y)    

        # "momentum" update with beta1
        vx = beta1*vx + (1-beta1)*dx
        vy = beta1*vy + (1-beta1)*dy
        # "rmsprop" update with beta2
        sx = beta2*sx + (1-beta2)*dx**2   
        sy = beta2*sy + (1-beta2)*dy**2
        
        # bias correction 
        vx = vx/(1-beta1**(epoch+1))
        vy = vy/(1-beta1**(epoch+1))
        
        sx = sx/(1-beta2**(epoch+1))
        sy = sy/(1-beta2**(epoch+1))

        x   = x - lrate*vx/(np.sqrt(sx)+eps)         
        y   = y - lrate*vy/(np.sqrt(sy)+eps)   
        
    
    log = np.asarray(log)
    return log
```

# Optimization

```python
# initial weights
x, y    = -1.5, 0.25
lrate   = 0.01
epoches = 500

log_bowl1       = plain_gd(x,y,bowl_der,lrate,epoches)
log_bowl2       = sgmoment(x,y,bowl_der,lrate,epoches,0.9)
log_bowl3       = rmsprop(x,y,bowl_der,lrate,epoches,(0.9,10e-8))
log_bowl4       = adam(x,y,bowl_der,lrate,epoches,(0.9,0.95,10e-8))
```

###  Plotting Bowl

```python
x = np.arange(-2, 2, 0.05)
y = np.arange(-1, 1, 0.05)

# plot surface
X, Y = np.meshgrid(x,y)
Z = bowl(X.ravel(), Y.ravel())
Z = Z.reshape(X.shape[0],-1)

fig = plt.figure(figsize=(12,9))
gs = fig.add_gridspec(1, 1)

ax1 = fig.add_subplot(gs[0, 0], projection='3d')
ax1.plot_surface(X,Y,Z, alpha=0.7,cmap=cm.bwr)
ax1.contour(X,Y,Z, 30, zdir='z', offset=-2, cmap=cm.bwr)
ax1.view_init(20, 60)
ax1.set_zlim(-2, 6)
```

```python
fig = plt.figure(figsize=(12,9))
gs = fig.add_gridspec(1, 1)

# plot optimization process
x_scatter = log_bowl1[:,0]
y_scatter = log_bowl1[:,1]
z_scatter = bowl(x_scatter, y_scatter)

x_scatter2 = log_bowl2[:,0]
y_scatter2 = log_bowl2[:,1]
z_scatter2 = bowl(x_scatter2, y_scatter2)

x_scatter3 = log_bowl3[:,0]
y_scatter3 = log_bowl3[:,1]
z_scatter3 = bowl(x_scatter3, y_scatter3)

x_scatter4 = log_bowl4[:,0]
y_scatter4 = log_bowl4[:,1]
z_scatter4 = bowl(x_scatter4, y_scatter4)

ax1 = fig.add_subplot(gs[0, 0], projection='3d')

ax1.plot(x_scatter,y_scatter,z_scatter,'ok-',label='plain GD')
ax1.plot(x_scatter2,y_scatter2,z_scatter2,'or-',label='SGD')
ax1.plot(x_scatter3,y_scatter3,z_scatter3,'og-',label='RMSprop')
ax1.plot(x_scatter4,y_scatter4,z_scatter4,'ob-',label='ADAM')

ax1.plot_surface(X,Y,Z, alpha=0.7,cmap=cm.bwr,zorder=2)
ax1.view_init(40, 60)
ax1.set_zlim(-2, 6)

plt.legend()
plt.show()
```
